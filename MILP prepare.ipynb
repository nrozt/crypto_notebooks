```python
from re import findall, sub

def make_real_polynom(expr):
    positive = findall(r'[xy]\d[^\']', expr)
    negative = findall(r'[xy]\d\'', expr)
    positive = [x[:-1] for x in positive]
    negative = ['%s + 1 - 2*%s' % (x[:-1], x[:-1]) for x in negative]
    return ' + '.join(positive + negative)

def transform_to_milp_exprs(cnf):
    sums = findall(r'\([^\(]+\)', cnf)
    sums = [x[1:-1] + ' ' for x in sums]
    new_sums = list()

    exprs = [make_real_polynom(x) + ' >= 1' for x in sums]
    return exprs

exprs = transform_to_milp_exprs(cnfs['Midori'])
exprs
```
